```python
import numpy as np
from sklearn.metrics import jaccard_score

result_dir = '/home/ganlu/Datasets/nclt/results/'
filepath = '/home/ganlu/Datasets/nclt/results/result_list.txt'

gt_all = np.array([])
pred_all = np.array([])
with open(filepath) as fp:
    for line in fp:
        print(line)
        
        result = np.loadtxt(result_dir + line[0:-1])
        gt = result[:,0]
        pred = result[:,1]
        print(gt.shape)
        print(pred.shape)
        gt_all = np.concatenate((gt_all, gt))
        pred_all = np.concatenate((pred_all, pred))
        
mask = (gt_all == 0)
pred_all[mask] = 0
mask = (pred_all == 0)
gt_all[mask] = 0

print(gt_all.shape)
print(pred_all.shape)
print( np.unique(np.concatenate((gt_all, pred_all), axis=0)) )
print(jaccard_score(gt_all, pred_all, average=None)) 


   
```
