```python
import numpy as np
from sklearn.metrics import jaccard_score

evaluation_dir = '/home/ganlu/la3dm_ws/src/la3dm/data/nclt/nclt_0429_evaluation_csm/'
result_list = evaluation_dir + 'result_list.txt'

gt_all = np.array([])
pred_all = np.array([])
with open(result_list) as fp:
    for line in fp:
        print(line)
        result = np.loadtxt(evaluation_dir + line[0:-1])
        if result.shape[0] == 0:
            continue
        gt = result[:,0]
        pred = result[:,1]
        gt_all = np.concatenate((gt_all, gt))
        pred_all = np.concatenate((pred_all, pred))

# Ignore background and sky label
pred_all = pred_all[gt_all != 0]
gt_all = gt_all[gt_all != 0]
pred_all = pred_all[gt_all != 13]
gt_all = gt_all[gt_all != 13]

        
print(gt_all.shape)
print(pred_all.shape)
print( np.unique(np.concatenate((gt_all, pred_all), axis=0)) )
print(jaccard_score(gt_all, pred_all, average=None)) 
```

```python
from sklearn.metrics import classification_report
print(classification_report(gt_all, pred_all))
```
